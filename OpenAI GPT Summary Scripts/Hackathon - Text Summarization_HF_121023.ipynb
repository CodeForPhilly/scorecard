```python
df["Relevant_Text"] = df["Relevant_Text"].apply(lambda x: clean(x,
    lower=True,
    no_punct=True,
    no_emoji=True,
    no_line_breaks=True,           # fully strip line breaks as opposed to only normalizing them
    no_numbers=True,               # replace all numbers with a special token
    replace_with_number="",
    lang="en"                       
    ))

df.head()
```

```python
import tensorflow as tf
gpu_options = tf.GPUOptions(allow_growth=True)
session = tf.InteractiveSession(config=tf.ConfigProto(gpu_options=gpu_options))
```

```python
#Text Summarization

# Initialize summarization pipeline
#summarization_pipeline = pipeline("summarization", model="t5-small", tokenizer="t5-small") # using bart in pytorch
summarization_pipeline = pipeline("summarization", model="t5-base", tokenizer="t5-base", framework="tf") # using t5 n tf
# Set parameters for summarization
max_length = 100
#min_length = 10

# Initialize an empty list to store summarized texts
summarized_texts = []
#df_out = df.copy()

# Iterate through the rows of df_out and perform summarization

for i in range(0, 50):
    request_explanation = df["Relevant_Text"][i]
    summarized_text = summarization_pipeline(request_explanation, max_length=max_length, num_beams=4, length_penalty=2.0)[0]["summary_text"]
    summarized_texts.append(summarized_text)


# Add the list of summarized texts to df_out as a new column

df["Summarized_Text"] = summarized_texts
```
